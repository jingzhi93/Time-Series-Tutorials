# Tutorials

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
```

## Loading and Transforming the Data

```python
raw = pd.read_csv("Index2018.csv")
df_comp = raw.copy()
df_comp.date = pd.to_datetime(raw.date, dayfirst=True)
df_comp.set_index("date", inplace=True)
df_comp = df_comp.asfreq('b')
df_comp = df_comp.fillna(method="ffill")
```

## Removing Surplus Data

```python
df_comp['market_value'] = df_comp.spx
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']

size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## White Noise

Condition of a white noise:

- Constant mean
- Constant variance
- No autocorrelation

```python
wn = np.random.normal(loc = df_comp.market_value.mean(), scale=df_comp.market_value.std(), size=len(df_comp))
```

```python
df_comp['wn'] = wn
```

```python
df_comp.wn.plot(figsize=(20,4))
plt.title("White Noise Time Series")
plt.show()
```

```python
df_comp.market_value.plot(figsize=(20,4))
plt.title("Market Value")
plt.ylim(0,3000)
plt.show()
```

## Random Walk

Equation of random walk:
$x_t = x_{t-1} + N(\mu, \sigma)$

```python
rw = pd.read_csv("RandWalk.csv")
rw.date = pd.to_datetime(rw.date, dayfirst=True)
rw.set_index("date", inplace=True)
rw = rw.asfreq('b')
rw = rw.fillna(method="ffill")
```

```python
df_comp['rw'] = rw
```

```python
df_comp.rw.plot(figsize=(20,4))
plt.title("Random Walk", size=24)
plt.show()
```

```python
df_comp.wn.plot(figsize=(20,4))
plt.title("White Noise Time Series")
plt.show()
```

## Stationarity

Stationarity of time series implies that taking consecutive samples of data with the sma size should have identical covariance regardless of starting point. Also known as "weak form" stationarity or "covariance" stationarity.
<div>
    <img src="img/cov.png">
</div>
Covariance stationaroty assumptions:
- Constant mean $\mu$
- Constant variance $\sigma^2$
- Constant covariance $Cov(x_n , x_{n+k}) = Cov(x_m , x_{m+k})$

Example
$Cov(x_1, x_4) = Cov(x_3, x_6)$

## ADF Test

$H_0$:Assumes non-stationary ( $\phi_1 < 1$) <br>
$H_1$: $\phi_1 = 1$ <br>

Test statistic < critical value => stationary

```python
results = sts.adfuller(df.market_value); results
```

```python
print(f"T-statistics: {results[0]}")
print(f"P-value: {results[1]}")
print(f"Number of lags used in the regression when determine the t-statistics: {results[2]}")
print(f"T-statistics level of significance: {results[4]}")
```

The t-statistics is greater than critical values of different significance value. Thus, we do not find sufficient evidence of stationarity in the dataset. <br>
The p-value = 0.41, this means the is a 40% chance of not rejecting the null, thus we do not confirm the data is stationary. <br>
The number of lags used in the regression when determine the t-statistics is 18 which means that there's some auto correlation going back 18 periods. <br>

```python
 sts.adfuller(wn)
```

```python
 sts.adfuller(rw)
```

## Correlation

### ACF 

ACF calculates the autocorrelation values for however many lags we are interested in simultaneously.
<div>
    <img src="img/acf.png" width="500"/>
</div>

Let's plot the ACF of the SnP. The x-axis shows the number of lags which is 40 and it is consistent to what we set in the function. The y-axis has the maximum value of 1.0 which makes sense since the max value of correlation is 1. <br>
The blue region of the graph shows the level of significance. If the value of the plot exceeds the blue region, it means that there's a significance in the autocorrelation. <br>
The greater the distance in time, the more unlikely it is that this autocorrelation persists. For instance, today's price are usually closer to the yesterday's prices than the price a month ago. Therefore we should make sure that the significance level at the higher lag to be significantly different from zero.<br>
From the graph, we can observe that autocorrelations barely diminishes even at the 40 periods backwards. This suggest that, prices even at a month back can still serves as a decent estimator.

```python
#generally we use lags = 40
sgt.plot_acf(df.market_value, lags=40, zero=False)
plt.title("ACF SnP")
plt.show()
```

Let's plot a ACF for white noise. We can see that most of the autocorrelation values are within the significance level which is the assumption of a white noise data, ie white noise does not have points that are autocorrelate with each others.

```python
sgt.plot_acf(wn, lags=40, zero=False)
plt.title("ACF White Noise")
plt.show()
```

```python
sgt.plot_acf(rw, lags=40, zero=False)
plt.title("Random Walk")
plt.show()
```

## PACF

PACF eliminates the indirect effects of lags on the data

#### ACF ####
Direct effects: $t_3 \rightarrow t_1$ <br>
Indirect effects: $t_3 \rightarrow t_2 \rightarrow t_1$

#### PACF ####
Direct effects: $t_3 \rightarrow t_1$ <br>

```python
sgt.plot_pacf(df.market_value, lags=40, zero=False)
plt.title("PACF market value")
plt.show()
```

```python
sgt.plot_pacf(wn, lags=40, zero=False)
plt.title("PACF white noise")
plt.show()
```

```python
sgt.plot_pacf(rw, lags=40, zero=False)
plt.title("PACF Random Walk")
plt.show()
```
